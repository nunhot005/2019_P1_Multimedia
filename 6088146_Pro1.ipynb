```python
def get_figsize(im):
    # What size does the figure need to be in inches to fit the image?
    dpi = plt.rcParams['figure.dpi']
    dim = im.shape
    figuresize = dim[1]/float(dpi), dim[0]/float(dpi)
    return(figuresize)

def display_image_actual_size_single(im_data):
    figuresize = get_figsize(im_data)
    # Create a figure of the right size with one axes that takes up the full figure
    fig = plt.figure(figsize=figuresize)
    # Add the single axis to fit the image to the screen boundary
    ax = fig.add_axes([0, 0, 1, 1])
    ax.imshow(im_data, cmap='gray')
    ax.axis('off')
    plt.show()
    return(fig)

def saltpepper_noise(image, prob):
    output = np.zeros(image.shape,np.uint8)
    thres = 1 - prob
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = np.random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn > thres:
                output[i][j] = 255
            else:
                output[i][j] = image[i][j]
    return output

```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

loadedImage = cv2.imread('image2.jpg')
image = cv2.cvtColor(loadedImage, cv2.COLOR_BGR2RGB)

fig = display_image_actual_size_single(image)
```

```python
weight = 100
sz = 23
prob  = 0.3
sobelSZ = 15

def manageImage(roi,i,j):

    if i == 0 and j == 0:
        return(roi)
    if i == 0 and j == 1:
        
        roiHalfSizeW = int(roi.shape[1] / 2)
        #mini 1
        roiLeft = roi[:, 0:roiHalfSizeW]
           
#         #bluegray
        roiLeft[:,:,0] =  roiLeft[:,:,2]
        roiLeft[:,:,1] =  roiLeft[:,:,2]
        
#         roiLeft = cv2.cvtColor(roiLeft, cv2.COLOR_RGB2GRAY)
#         roiLeft = cv2.cvtColor(roiLeft, cv2.COLOR_GRAY2RGB)
#         plt.imshow(roiLeft, cmap='gray')
#         roiLeft = cv2.applyColorMap(roiLeft, cv2.COLORMAP_JET)
        
        #mini 2
        roiRight = roi[:, roiHalfSizeW:2*roiHalfSizeW]
        #twocolorGR
        roiRight[:,:,0] = 0
        roiRight = cv2.cvtColor(roiRight,cv2.COLOR_BGR2RGB)
    
        miniMerge = np.concatenate((roiLeft,roiRight),axis = 1)
        roi = miniMerge
        
    if i == 0 and j == 2:
        onecolorR = roi.copy()
        onecolorR[:,:,0] = 0
        onecolorR[:,:,1] = 0
        onecolorR = cv2.cvtColor(onecolorR,cv2.COLOR_BGR2RGB)
        roi = onecolorR
    if i == 0 and j == 3:
        twocolorBG = roi.copy()
        twocolorBG[:,:,2] = 0
        twocolorBG = cv2.cvtColor(twocolorBG,cv2.COLOR_BGR2RGB)
        roi = twocolorBG
    if i == 1 and j == 0:
        brighter = cv2.add(roi,weight)
        brighter = cv2.cvtColor(brighter, cv2.COLOR_BGR2RGB)
        roi = brighter
    if i == 1 and j == 1:
        medianFiltered = cv2.medianBlur(roi,sz)
        medianFiltered = cv2.cvtColor(medianFiltered,cv2.COLOR_BGR2RGB)
        roi = medianFiltered
    if i == 1 and j == 2:
        negative = cv2.subtract(255,roi)
        negative = cv2.cvtColor(negative, cv2.COLOR_BGR2RGB)
        roi = negative
    if i == 1 and j == 3:
        sobelEdge = cv2.Sobel(roi, -1, 1, 1, ksize=sobelSZ)    #mix v and h
        sobelEdge = cv2.cvtColor(sobelEdge,cv2.COLOR_BGR2RGB)
        roi = sobelEdge
    if i == 2 and j == 0:    
        blurGauss = cv2.GaussianBlur(roi, (sz,sz), 0)
        blurGauss = cv2.cvtColor(blurGauss,cv2.COLOR_BGR2RGB)
        roi = blurGauss
    if i == 2 and j == 1:
        imSP = saltpepper_noise(roi,prob) 
        imSP = cv2.cvtColor(imSP,cv2.COLOR_BGR2RGB)
        roi = imSP
    if i == 2 and j == 2:
        blurBox = cv2.boxFilter(roi, -1, (sz,sz))
        blurBox = cv2.cvtColor(blurBox,cv2.COLOR_BGR2RGB)
        roi = blurBox
    if i == 2 and j == 3:
        higher = cv2.multiply(roi,2)
        higher = cv2.cvtColor(higher, cv2.COLOR_BGR2RGB)
        roi = higher
    return(roi)
```

```python
roiH = int(image.shape[0]/3)
roiW = int(image.shape[1]/4)

horizon = []
vertical = []

height = 0
width = 0

# The following logic is the result of consulting with
# a section 1 friend named Tanawin Wichit, 6088221.
for i in range(0,3) :
    for j in range(0,4):
        height = i*roiH
        width = j*roiW
        roi = image[height:height+roiH,width:width+roiW]
        roi = manageImage(roi,i,j)
        horizon.append(roi)
       
        if j == 3:
            mergeHor = np.concatenate(horizon,axis=1)
            horizon = []
            vertical.append(mergeHor)
            mergeVer = np.concatenate(vertical,axis=0)
            
trueim = display_image_actual_size_single(mergeVer)
trueim.savefig("6088146_Pro1.jpg", bbox_inches = 0, pad_inches = 0)
```
